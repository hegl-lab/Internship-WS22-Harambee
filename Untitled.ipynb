```python
def get_sentencetransformer_embeddings(df, unique=False):
    ## transforms sentence into sentence embeddings
    model_senetenceBert = SentenceTransformer('bert-base-uncased')
    model_senetenceBert
    model_senetenceBert.max_seq_length = 32
    if(unique == False):
        embedding_1 = model_senetenceBert.encode(df.job_title.tolist())
    if(unique == True):
        embedding_1 = model_senetenceBert.encode(df.job_title.unique().tolist())
    print(embedding_1.shape) ## shape of (100, 768), its like a matrix with 100 rows, where each row is a sentence 
                         ## and the context of each sentence is embedded into 768 values

    # feed the sentence embedding into the clustering model
    
    return embedding_1
#embedding_total = get_sentencetransformer_embeddings(df)
embedding_total_unique = get_sentencetransformer_embeddings(df,unique=True)
```

```python
def plot_tnse(df, embeddings, n_clusters_ = 5):
    from sklearn.cluster import AgglomerativeClustering    
    from sklearn.manifold import TSNE

    hc = AgglomerativeClustering(n_clusters = n_clusters_, affinity = 'euclidean',linkage ='ward') #, 'ward'
    hc = hc.fit_predict(embeddings)
    def_of_embedded_part_ = df#.head(1000)
    def_of_embedded_part_['clusters'] = list(hc)

    tsne = TSNE(n_components=2, verbose=1, random_state=123)
    df_temp = pd.DataFrame()
    z = tsne.fit_transform(embeddings)
    df_temp["comp-1"] = z[:,0]
    df_temp["comp-2"] = z[:,1]
    
    sns.scatterplot(x="comp-1", y="comp-2", hue= def_of_embedded_part_['clusters'],
                palette=sns.color_palette("hls", n_clusters_),
                data=df_temp).set(title="Job titles T-SNE projection") #hue=df.y.tolist(),
```
